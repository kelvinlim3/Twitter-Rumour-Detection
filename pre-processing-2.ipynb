```python
import pandas as pd
import numpy as np
import re
import nltk
from textblob import TextBlob
```

```python
def pre_process_text(df):
    text = list(df['text'])
    sentiment = []
    for i in range(len(text)):
        text[i] = " ".join(filter(lambda x:x[0]!='@', text[i].split()))
        text[i] = " ".join(filter(lambda x:x[0:4]!='http', text[i].split()))
        text[i] = " ".join(filter(lambda x:x[0:4]!='&amp', text[i].split()))
        text[i] = re.sub(r'[^a-zA-Z ]','',text[i])
        text[i] = " ".join(filter(lambda x:x[0:1]!='Q', text[i].split()))
        text[i] = " ".join(filter(lambda x:x[0:1]!='A', text[i].split()))
        sentiment.append(TextBlob(text[i]).sentiment.polarity)
    return text, sentiment
```

```python
sources = pd.read_csv('train_sources.csv', index_col=0)
sources['text'], sources['polarity'] = pre_process_text(sources)
sources.fillna('', inplace=True)
sources.text = sources.text + " " + sources.hashtag
sources.drop_duplicates(subset=['text'][0:100],inplace=True)
sources
```

```python
replies = pd.read_csv('train_replies.csv', index_col=0)
replies['text'], replies['polarity'] = pre_process_text(replies)
replies.fillna('', inplace=True)
replies.text = replies.text + " " + replies.hashtag
replies.drop_duplicates(subset=['text'][0:100],inplace=True)
replies
```

```python
import nltk
from nltk.tokenize import TweetTokenizer
from nltk.corpus import stopwords, wordnet

tt = TweetTokenizer()
stopwords = set(stopwords.words('english')) #note: stopwords are all in lowercase
lemmatizer = nltk.stem.wordnet.WordNetLemmatizer()

def lemmatize(word):
    lemma = lemmatizer.lemmatize(word,'v')
    if lemma == word:
        lemma = lemmatizer.lemmatize(word,'n')
    return lemma

sources_tokens_list = []
for _id, row in sources.iterrows():
    text = row['text']
    # tokenize tweet
    tokens = tt.tokenize(text)
    # convert to lowercase
    tokens = [tok.lower() for tok in tokens]
    # remove stopwords
    tokens = [tok for tok in tokens if tok not in stopwords]
    # lemmatize
    tokens = [lemmatize(tok) for tok in tokens]
    
    sources_tokens_list.append(tokens)

replies_tokens_list = []
for _id, row in replies.iterrows():
    text = row['text']
    # tokenize tweet
    tokens = tt.tokenize(text)
    # convert to lowercase
    tokens = [tok.lower() for tok in tokens]
    # remove stopwords
    tokens = [tok for tok in tokens if tok not in stopwords]
    # lemmatize
    tokens = [lemmatize(tok) for tok in tokens]
    
    replies_tokens_list.append(tokens)
    
replies.head()
```

```python
sources.insert(4, 'tokens', sources_tokens_list)
sources.head()
```

```python
replies.insert(4, 'tokens', replies_tokens_list)
replies.head()
```

```python
def encode_df(df):
    encoded = df
    encoded['verified']=encoded['verified'].map({True:1, False:0})
    if 'source class' in list(df.columns):
        encoded['source class']=encoded['source class'].map({'rumour':1, 'nonrumour':0})
    else:
        encoded['class']=encoded['class'].map({'rumour':1, 'nonrumour':0})
        
    followers_encoded = []
    for n in list(encoded.followers):
        if n <1000:
            followers_encoded.append(1)
        elif n<5000:
            followers_encoded.append(2)
        elif n<20000:
            followers_encoded.append(3)
        elif n<50000:
            followers_encoded.append(4)
        elif n<200000:
            followers_encoded.append(5)
        else:
            followers_encoded.append(6)
        
    encoded.followers = followers_encoded
    
    return encoded
```

```python
sources = encode_df(sources)
sources.head()
```

```python
replies = encode_df(replies)
replies.head()
```

```python
replies.drop(['user','text','hashtag','source id'],axis=1,inplace=True)
replies
```

```python
sources.drop(['user','text','hashtag'],axis=1,inplace=True)
sources
```

```python
def count_tokens(df):
    token_counts = {}
    for token_list in list(df['tokens']):
        for token in token_list:
            if token in token_counts:
                token_counts[token] += 1
            else:
                token_counts[token] = 1
    return {k: v for k, v in sorted(token_counts.items(), key=lambda item: item[1],reverse=True)}
            
replies_uni_counts = count_tokens(replies)
replies_uni_counts
```

```python
new_cols = [x for x in list(replies.columns) if x != 'source class']+list(replies_uni_counts.keys())[0:500]+['source class']
replies = replies.reindex(columns=new_cols, fill_value=0)
replies.head()
```

```python
for i, row in replies.iterrows():
    for token in row['tokens']:
        if token in list(replies.columns):
            replies.at[i, token] += 1
replies.drop(['tokens'],axis=1,inplace=True)
replies
```
